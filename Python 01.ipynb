```python
from turtle import *
from math import sin, cos, pi

tracer(0, 0)
speed(0)

R = 0.5

for j in range(360):
    r = R*cos(j/180*pi)
    g = R*sin(j/180*pi)*cos(j/180*pi)
    b = R*sin(j/180*pi)*sin(j/180*pi)
    
    r = (r+1)/2
    g = (g+1)/2
    b = (b+1)/2
    
    color(r,g,b)
    for i in range(4):
        fd(100)
        rt(90)
    rt(1)
done()
```
